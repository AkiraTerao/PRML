```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate
from scipy.stats import uniform, norm

# 被積分関数
f = lambda x: norm.pdf(x)
h = lambda x: np.exp(-(x-3)**2/2) + np.exp(-(x-6)**2/2)
y = lambda x: h(x) * f(x)

# グラフ描画
ix = np.arange(-10, 10, 0.01)
plt.plot(ix, f(ix))
plt.plot(ix, h(ix))
plt.plot(ix, y(ix))
plt.show()

# scipy.integrateでの積分
I = scipy.integrate.quad(y, -np.inf, np.inf)[0]
print I

N = 100

# 通常のモンテカルロ積分
x = norm.rvs(size=N)
I = np.mean(h(x))
print I

# 重点サンプリングによるモンテカルロ積分
a, b = 1, 8
u = uniform(loc=a, scale=b-a)
g = lambda x: u.pdf(x)
x = u.rvs(size=N)
I = np.mean((f(x)/g(x))*h(x))
print I

ix = np.arange(-10, 10, 0.01)
plt.plot(ix, g(ix))
plt.show()
```
