```python
# 同じことをscipy.statsで実現
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt

# 一様分布に従う確率分布からランダムサンプリング
np.random.seed()
N = 10000
# [0.0, 1.0]の一様分布に従う確率変数
rv = uniform(loc=0.0, scale=1.0)
# 一様分布からサンプリング
x = rv.rvs(size=N)
nbins = 50
plt.hist(x, nbins, normed=True)

# 真のPDFを描画
x = np.linspace(rv.ppf(0), rv.ppf(1), 100)
plt.plot(x, uniform.pdf(x), 'r-', lw=2, label='uniform pdf')

plt.show()
```

```python
# 同じことをscipy.statsで実現
# 確率変数オブジェクトを使わないで
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt

# 一様分布に従う確率分布からランダムサンプリング
np.random.seed()
N = 10000
loc = 0.0
scale = 1.0

# 一様分布からサンプリング
x = uniform.rvs(loc=loc, scale=scale, size=N)
nbins = 50
plt.hist(x, nbins, normed=True)

# 真のPDFを描画
x = np.linspace(uniform.ppf(0, loc=loc, scale=scale), uniform.ppf(1, loc=loc, scale=scale), 100)
plt.plot(x, uniform.pdf(x, loc=loc, scale=scale), 'r-', lw=2, label='uniform pdf')

plt.show()
```

```python
# 逆変換法で一様分布から指数分布を得る
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

nbins = 50

# 指数分布のパラメータ
scale = 1.0

# 逆変換法で一様乱数から指数分布の乱数を得る
np.random.seed()
N = 100000
U = scipy.stats.uniform(loc=0.0, scale=1.0).rvs(size=N)

# 指数分布の累積分布関数の逆関数を用いて変換
X1 = - scale * np.log(1 - U)

# 生成した元の一様乱数を描画
plt.figure(1)
plt.hist(U, nbins, normed=True)

# 変換した指数分布の乱数と理想的なPDFを描画
plt.figure(2)
rv = scipy.stats.expon(scale=scale)
plt.hist(X1, nbins, normed=True)
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)
y = rv.pdf(x)
plt.plot(x, y, 'r-', lw=2)
plt.xlim((rv.ppf(0.01), rv.ppf(0.99)))

# scipyの乱数生成関数を利用した場合
plt.figure(3)
rv = scipy.stats.expon(scale=scale)
X2 = rv.rvs(N)
plt.hist(X2, nbins, normed=True)

x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 1000)
y = rv.pdf(x)
plt.plot(x, y, 'r-', lw=2)
plt.xlim((rv.ppf(0.01), rv.ppf(0.99)))

plt.show()
```

```python
# 逆変換法で一様分布からロジスティック分布を得る
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

nbins = 50

# 逆変換法で一様分布からロジスティック分布を得る
np.random.seed()
N = 10000
U = np.random.uniform(0.0, 1.0, N)
X1 = - np.log(1 / U - 1)

# 生成した元の一様分布を描画
plt.figure(1)
plt.hist(U, nbins, normed=True)

# 変換したロジスティック分布と理想的なPDFを描画
plt.figure(2)
plt.xlim((-10, 10))
plt.hist(X1, nbins, normed=True)

x = np.linspace(-10, 10, 1000)
y = scipy.stats.logistic.pdf(x)
plt.plot(x, y, 'r-', lw=2)

# numpyのロジスティック分布に従う乱数生成関数を利用した場合
plt.figure(3)
plt.xlim((-10, 10))
X2 = np.random.logistic(size=N)
plt.hist(X2, nbins, normed=True)

x = np.linspace(-10, 10, 1000)
y = scipy.stats.logistic.pdf(x)
plt.plot(x, y, 'r-', lw=2)

plt.show()
```

```python
# 逆変換法で一様分布からコーシー分布を得る
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

nbins = 50

# 逆変換法で一様分布からコーシー分布を得る
# コーシー分布は裾野が広く極端な値も出やすいためplotするときは
# それらのデータを切り捨てている
np.random.seed()
N = 10000
U = np.random.uniform(0.0, 1.0, N)
X1 = np.tan(np.pi * (U - 0.5))
X1 = X1[(X1>-20) & (X1<20)]

# 生成した元の一様分布を描画
plt.figure(1)
plt.hist(U, nbins, normed=True)

# 変換したコーシー分布と理想的なPDFを描画
plt.figure(2)
plt.xlim((-20, 20))
plt.hist(X1, nbins, normed=True)

x = np.linspace(-20, 20, 1000)
y = scipy.stats.cauchy.pdf(x)
plt.plot(x, y, 'r-', lw=2)

# numpyのコーシー分布に従う乱数生成関数を利用した場合
plt.figure(3)
plt.xlim((-20, 20))
X2 = np.random.standard_cauchy(N)
X2 = X2[(X2>-20) & (X2<20)]
plt.hist(X2, nbins, normed=True)

x = np.linspace(-20, 20, 1000)
y = scipy.stats.cauchy.pdf(x)
plt.plot(x, y, 'r-', lw=2)

plt.show()
```

```python
# 逆変換法で一様分布からガンマ分布を得る
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import scipy.special

nbins = 50

# ガンマ分布のパラメータ
k = 2.0      # shape
theta = 2.0  # scale

# 逆変換法で一様分布からガンマ分布を得る
np.random.seed()
N = 10000
U = np.random.uniform(0.0, 1.0, N)
X1 = theta * scipy.special.gammaincinv(k, U * scipy.special.gamma(k))

# 生成した元の一様分布を描画
plt.figure(1)
plt.hist(U, nbins, normed=True)

# 変換したガンマ分布と理想的なPDFを描画
plt.figure(2)
plt.xlim((0, 20))
plt.hist(X1, nbins, normed=True)

x = np.linspace(0, 20, 1000)
y = scipy.stats.gamma(a=k, scale=theta).pdf(x)
plt.plot(x, y, 'r-', lw=2)

# numpyのガンマ分布に従う乱数生成関数を利用した場合
plt.figure(3)
plt.xlim((0, 20))
X2 = np.random.gamma(shape=k, scale=theta, size=N)
plt.hist(X2, nbins, normed=True)

x = np.linspace(0, 20, 1000)
y = scipy.stats.gamma(a=k, scale=theta).pdf(x)
plt.plot(x, y, 'r-', lw=2)

plt.show()
```

```python
# 逆変換法で一様分布からベータ分布を得る
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import scipy.special

nbins = 50

# ベータ分布のパラメータ
# alpha = 0.5
# beta  = 0.5

alpha = 5.0
beta = 1.0

# 逆変換法で一様分布からガンマ分布を得る
np.random.seed()
N = 10000
U = np.random.uniform(0.0, 1.0, N)
X1 = scipy.special.betaincinv(alpha, beta, U)

# 生成した元の一様分布を描画
plt.figure(1)
plt.hist(U, nbins, normed=True)

# 変換したベータ分布と理想的なPDFを描画
plt.figure(2)
plt.xlim((0, 1))
plt.hist(X1, nbins, normed=True)

x = np.linspace(0, 1, 1000)
y = scipy.stats.beta.pdf(x, alpha, beta)
plt.plot(x, y, 'r-', lw=2)

# numpyのベータ分布に従う乱数生成関数を利用した場合
plt.figure(3)
plt.xlim((0, 1))
X2 = np.random.beta(alpha, beta, size=N)
plt.hist(X2, nbins, normed=True)

x = np.linspace(0, 1, 1000)
y = scipy.stats.beta.pdf(x, alpha, beta)
plt.plot(x, y, 'r-', lw=2)

plt.show()
```
