```python
# 逆変換法で一様分布から指数分布を得る
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

nbins = 50

# 指数分布のパラメータ
lam = 1.0
scale = 1.0 / lam

# 逆変換法で一様分布から指数分布を得る
np.random.seed()
N = 10000
U = np.random.uniform(0.0, 1.0, N)
X1 = - scale * np.log(1 - U)

# 生成した元の一様分布を描画
plt.figure(1)
plt.hist(U, nbins, normed=True)

# 変換した指数分布と理想的なPDFを描画
plt.figure(2)
plt.xlim((0, 10))
plt.hist(X1, nbins, normed=True)

x = np.linspace(0, 10, 1000)
y = scipy.stats.expon.pdf(x, scale=scale)
plt.plot(x, y, 'r-', lw=2)

# numpyの指数分布に従う乱数生成関数を利用した場合
plt.figure(3)
plt.xlim((0, 10))
X2 = np.random.exponential(scale=scale, size=N)
plt.hist(X2, nbins, normed=True)

x = np.linspace(0, 10, 1000)
y = scipy.stats.expon.pdf(x, scale=scale)
plt.plot(x, y, 'r-', lw=2)
```

```python
# 逆変換法で一様分布からロジスティック分布を得る
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

nbins = 50

# 逆変換法で一様分布からロジスティック分布を得る
np.random.seed()
N = 10000
U = np.random.uniform(0.0, 1.0, N)
X1 = - np.log(1 / U - 1)

# 生成した元の一様分布を描画
plt.figure(1)
plt.hist(U, nbins, normed=True)

# 変換したロジスティック分布と理想的なPDFを描画
plt.figure(2)
plt.xlim((-10, 10))
plt.hist(X1, nbins, normed=True)

x = np.linspace(-10, 10, 1000)
y = scipy.stats.logistic.pdf(x)
plt.plot(x, y, 'r-', lw=2)

# numpyのロジスティック分布に従う乱数生成関数を利用した場合
plt.figure(3)
plt.xlim((-10, 10))
X2 = np.random.logistic(size=N)
plt.hist(X2, nbins, normed=True)

x = np.linspace(-10, 10, 1000)
y = scipy.stats.logistic.pdf(x)
plt.plot(x, y, 'r-', lw=2)
```

```python
# 逆変換法で一様分布からコーシー分布を得る
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

nbins = 50

# 逆変換法で一様分布からコーシー分布を得る
# コーシー分布は裾野が広く極端な値も出やすいためplotするときは
# それらのデータを切り捨てている
np.random.seed()
N = 10000
U = np.random.uniform(0.0, 1.0, N)
X1 = np.tan(np.pi * (U - 0.5))
X1 = X1[(X1>-20) & (X1<20)]

# 生成した元の一様分布を描画
plt.figure(1)
plt.hist(U, nbins, normed=True)

# 変換したコーシー分布と理想的なPDFを描画
plt.figure(2)
plt.xlim((-20, 20))
plt.hist(X1, nbins, normed=True)

x = np.linspace(-20, 20, 1000)
y = scipy.stats.cauchy.pdf(x)
plt.plot(x, y, 'r-', lw=2)

# numpyのコーシー分布に従う乱数生成関数を利用した場合
plt.figure(3)
plt.xlim((-20, 20))
X2 = np.random.standard_cauchy(N)
X2 = X2[(X2>-20) & (X2<20)]
plt.hist(X2, nbins, normed=True)

x = np.linspace(-20, 20, 1000)
y = scipy.stats.cauchy.pdf(x)
plt.plot(x, y, 'r-', lw=2)
```
